```python
import pandas as pd

training_loss_all = pd.read_json('./saved_0624/saved_losses/training_loss_epoch_0.json', orient='index')
for i in range(1, 926):
    training_loss_all = pd.concat([training_loss_all, 
                                   pd.read_json(f'saved_0624/saved_losses/training_loss_epoch_{i}.json', orient='index')])

val_loss_all = pd.read_json('saved_0624/saved_losses/val_loss_epoch_0.json', orient='index')
for i in range(1, 926):
    val_loss_all = pd.concat([val_loss_all, 
                              pd.read_json(f'saved_0624/saved_losses/val_loss_epoch_{i}.json', orient='index')])

training_loss_all.columns = ['loss']
val_loss_all.columns = ['loss']
```

```python
import re
training_loss_all['epoch'] = [re.search("Epoch +(\d+) +Step +(\d+)",x).group(1) for x in list(training_loss_all.index.values)]
# training_loss_all['step'] = [re.search("Epoch +(\d+) +Step +(\d+)",x).group(2) for x in list(training_loss_all.index.values)]
val_loss_all['epoch'] = [re.search("Epoch +(\d+) +Step +(\d+)",x).group(1) for x in list(val_loss_all.index.values)]
# val_loss_all['step'] = [re.search("Epoch +(\d+) +Step +(\d+)",x).group(2) for x in list(val_loss_all.index.values)]
training_loss_all['step'] = training_loss_all.reset_index().index
val_loss_all['step'] = val_loss_all.reset_index().index
```

```python
import numpy as np
training_loss_all['log_loss'] = np.log(training_loss_all.loss)
val_loss_all['log_loss'] = np.log(val_loss_all.loss)
training_loss_all
```

```python
import matplotlib.pyplot as plt
import matplotlib.collections as collections
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(20,5))
ax1.plot(training_loss_all.step.values, training_loss_all.loss.values, color='blue', linewidth=1)
ax1.set_title('training loss')
ax2.plot(training_loss_all.step, training_loss_all.log_loss, color='blue', linewidth=1)
ax2.set_title('training log loss')
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(20,5))
ax1.plot(val_loss_all.step.values, val_loss_all.loss.values, color='blue', linewidth=1)
ax1.set_title('validation loss')
ax2.plot(val_loss_all.step, val_loss_all.log_loss, color='blue', linewidth=1)
ax2.set_title('validation log loss')
plt.show()
```
